# Python | Pandas Visualizations
Pandas is one of the most popular Python packages used in data science. Pandas offer a powerful, and flexible data structure ( Dataframe & Series ) to manipulate, and analyze the data. Visualization is the best way to interpret the data. 

Python has many popular plotting libraries that make visualization easy. Some of them are matplotlib, seaborn, and plotly. It has great integration with matplotlib. We can plot a dataframe using the plot() method. 

```python
# importing required library
import pandas as pd
import matplotlib.pyplot as plt
  
# A dictionary which represents data
dict = { 'name':['n1','n2','n3','n4','n5','n6'],
              'age':[20,20,21,20,21,20],
              'math_marks':[100,90,91,98,92,95],
              'physics_marks':[90,100,91,92,98,95],
              'chem_marks' :[93,89,99,92,94,92]
              }
  
# creating a data frame object
df = pd.DataFrame(dict)
  
# show the dataframe
# by default head() show first five rows from top
df.head()
```

# Plots
There are a number of plots available to interpret the data. Each graph is used for a purpose. Some of the plots are BarPlots, ScatterPlots, and Histograms, etc.



# 1.Scatter Plot
To get the scatterplot of a dataframe all we have to do is to just call the plot() method by specifying some parameters. 
Specify that you want a scatter plot with the kind argument:
kind = 'scatter'
A scatter plot needs an x- and a y-axis.
Create a scatter plot using plt.scatter()

```python
# scatter plot
df.plot(kind = 'scatter',
        x = 'math_marks',
        y = 'physics_marks',
        color = 'green')
  
# set the title
plt.title('ScatterPlot')
  
# show the plot
plt.show()
```

#Marker Shapes
The matplotlib markers module in python provides all the functions to handle markers.
By default it is always a circle shaped marker.
square="s", diamond="D", star="*", circle="o"

```python
import matplotlib.pyplot as plt

price = [2.50, 1.23, 4.02, 3.25, 5.00, 4.40]
sales_per_day = [34, 62, 49, 22, 13, 19]

#marker changes the shape of lable
plt.scatter(price, sales_per_day, marker="D",)

plt.xlabel("price")
plt.ylabel("sales_per_day")
plt.show()
```

# 2.Bar Plot 
Similarly, we have to specify some parameters for plot() method to get the bar plot. 

```python
# bar plot
df.plot(kind = 'bar',
        x = 'name',
        y = 'physics_marks',
        color = 'green')
  
# set the title
plt.title('BarPlot')
  
# show the plot
plt.show()
```

#Stacked bar plot
stacked= true

```python
df.plot(kind = 'bar',
        x = 'name',
        y = 'physics marks')

df.plot.bar(stacked=True, title='Students Marks')
    
plt.title('BarPlot')

plt.show()
```

# 3.Line Plot 
The line plot of a single column is not always useful, to get more insights we have to plot multiple columns on the same graph. To do so we have to reuse the axes. 

```python
#Get current axis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
ax = plt.gca() 
  
# line plot for math marks
df.plot(kind = 'line',
        x = 'name',
        y = 'math_marks',
        color = 'green',ax = ax)
  
# line plot for physics marks
df.plot(kind = 'line',x = 'name',
        y = 'physics_marks',
        color = 'blue',ax = ax)
  
# line plot for chemistry marks
df.plot(kind = 'line',x = 'name',
        y = 'chem_marks',
        color = 'black',ax = ax)
  
# set the title
plt.title('LinePlots')
  
# show the plot
plt.show()
```

# 4.Area Plot
This is an extended version of the line plot. In the area plot, instead of leaving the plot to connected data points, the area under the line is filled with colors. This is helpful in cases when you want to show the proportion of values captured by a particular value. As you can plot multiple variables, this can give you insights about when the variables are overlapping.

```python
# defining dataframe 
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(20, 4),
       columns =['A', 'B', 'C', 'D'])

# area plot 
df.plot.area()
```

# 5.Barh plot
Barh allows you to swap the axes. Therefore, you will have y-axes as the labels/categories and the x-axes will represent the values.

```python
# defining the dataframe
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10, 4), columns = ['A','B','C','D'])

# barh ploting
df.sum().plot.barh() 
```

# 6.Density or KDE Plot 
This plot visualizes the probability density function of a continuous random variable. This doesn’t directly tell the probability of the value taken by a random variable. In fact, for a continuous random variable, the probability of any value is zero and we are only concerned about the probabilities of a range of values. It only gives how many data points may be present around specified values.

```python
# defining the Series of random numbers
import pandas as pd
import numpy as np
df = pd.Series(np.random.rand(200)) 

# density plotting
df.plot.kde()

# Note: Density plot is only applicable for pandas Series 
```

# 7.Histogram Plot 
Histograms are used to represent the frequency of numerical variables. These are subversions of bar plots with the changes that in the histogram, we talk about numerical values. There are no categories but the numeric data is divided among small buckets called bins. These bins take in the number of values that fall in the range of the bin. Histograms are also quoted as frequency polygons when the bars are replaced by connecting lines from the midpoint of the bars.

```python
# defining a dataframe
import pandas as pd
import numpy as np
df = pd.DataFrame(
        { 
            "a": np.sqrt(np.random.randn(100) +1),
            "b": np.random.randn(100),
        },
        columns = ["a","b"],
    )

# histogram plotting
df.plot.hist(alpha = 0.5)

# Here, the “alpha” parameter defines the transparency of colors for the histograms.
```

# 8.Box Plot  
Box plots are very crucial and important plots. These plots help in understanding the overall trend and spread of a feature of a dataset. For numerical data, it represents where 50% of data lies, where the median of the data lies, and it also specifies the boundary conditions in the form of whiskers.

```python
# defining a dataframe
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(30,4),columns=["A","B","C","D"])

# box plotting
df.plot.box()
```

# 9.Hexbin Plot 
These plots can be alternative to scatter plots when the data is very dense. They represent data in the form of a honeycomb, the hexagons in the plot. It is also called a hexagon binning plot as it combines the power of bins in the histogram in the form of a hexagon. The color of each hexbin defines the number of points in that range.

```python
# defining a dataframe
df = pd.DataFrame(np.random.randn(100, 2), columns=["a", "b"])

df["b"] = df["b"] + np.arange(100) # it will reaarage the values from 1 to 100 and it will be stored in b

# hexbin plotting
df.plot.hexbin(x="a", y="b", gridsize=25);

# Here, the grid size controls the number of hexagons in the x-direction. By default, it is 100.
```

# 10.Pie Plot
Pie plots are used to represent the portion of a value as compared to others in a whole. They represent how much percent of the circular area a value is contributing and fill it with color. 

```python
# defining the pandas series
# pie plot is allowed only for pandas series
series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="Pie plot")

# pie plotting
series.plot.pie(figsize=(6, 6))

# The figsize parameter is used to control the figure size of the plot
```

#explode
The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```
